```python
# Plotting ROC-Curve 

def plotROCCurve(yPredicted, predScore, yTrue, figSize = (8,8), saveFigure = False, filePath = None):
    
    #logging.info("Plotting ROC Curve")
    logit_roc_auc = roc_auc_score(yTrue, yPredicted)
    fpr, tpr, thresholds = roc_curve(yTrue, predScore)
    print()
    plt.figure(figsize = figSize)
    plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([-0.01, 1.01])
    plt.ylim([-0.01, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Plot', y = 1.03, fontsize = 18)
    plt.legend(loc="lower right", fontsize = 'x-large')
    if not saveFigure:
        plt.show()
    else:
        try:
            plt.savefig(filePath)
        except Exception as e:
            print('Exception encountered while saving the plot to file. Make sure the file path inlcuding file name has proper image extension ')
            print('ValueError: ', str(e))
        
    plt.close()
```

```python
def getGainsTable(trueValue, predictedScore, buckets = 10, toExcel = False):
    
    #logging.info("Building Gains Chart")
    df=pd.DataFrame({"Label":trueValue,"modelScore":predictedScore})
    riskScores=df.sort_values(['modelScore'], ascending=[False])
    lableList=riskScores['Label'].tolist()
    decile=list(range(1, buckets+1))
    withoutModel=[sum (i) for i in np.array_split(np.array( [1]*sum(lableList) ), buckets)]
    prospects=[len (i) for i in np.array_split(np.array( lableList ), buckets)]
    withModel=[sum (i) for i in np.array_split(np.array( lableList ), buckets)]          
    gainsChart=pd.DataFrame({'Decile':decile,'prospects':prospects,'withModel':withModel,'withoutModel':withoutModel})
    gainsChart["lift"]=np.around(((gainsChart["withModel"]-gainsChart["withoutModel"])/(gainsChart["withoutModel"])*100))
    gainsChart["withModelRecall"]=np.around(((np.cumsum(gainsChart["withModel"])/np.sum(gainsChart["withModel"]))*100),2)
    gainsChart["withoutModelRecall"]=np.around(((np.cumsum(gainsChart["withoutModel"])/np.sum(gainsChart["withoutModel"]))*100))
    gainsChart["binTargetRate"]=np.round((gainsChart["withModel"]/gainsChart["prospects"])*100,decimals=2)
    gainsChart["cumulativeTargetRate"]=np.round((np.cumsum(gainsChart["withModel"])/np.cumsum(gainsChart["prospects"]))*100,decimals=2)
    gainsChart["binLift"]=np.around(((gainsChart["withModelRecall"]-gainsChart["withoutModelRecall"])/(gainsChart["withoutModelRecall"])*100),2)
    if toExcel:
        gainsChart[['lift','withModelRecall','withoutModelRecall','binTargetRate','cumulativeTargetRate','binLift']] = gainsChart[['lift','withModelRecall','withoutModelRecall','binTargetRate','cumulativeTargetRate','binLift']].applymap(lambda x: str(x)+'%') 
    return gainsChart
```
