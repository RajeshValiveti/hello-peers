```python
from sklearn import metrics
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
```

```python
data3hot.columns
```

```python
x=data3hot.drop(['Experience_Yrs',
       'Percentile_ET', 'Percent_MBA','Course_Arts', 'Course_Commerce',
       'Course_Computer_App', 'Course_Engineering', 'Course_Management',
       'Course_Others', 'Course_Science','Entrance_CAT', 'Entrance_GMAT',
       'Entrance_GSAT', 'Entrance_GCET', 'Entrance_KMAT', 'Entrance_MAT',
       'Entrance_MNG', 'Entrance_PGCET', 'Entrance_XAT', 'Spec_Finance',
       'Spec_HR', 'Spec_IB'],axis=1)
y=data3hot[['Percent_MBA']]
```

```python
kf=KFold(n_splits=3,shuffle=True,random_state=2)
rmse=[]
r2=[]
for train,test in kf.split(x,y):
    LR=LinearRegression()
    Xtrain,Xtest=x.iloc[train,:],x.iloc[test,:]
    Ytrain,Ytest=y.iloc[train],y.iloc[test]
    LR.fit(Xtrain,Ytrain)
    Y_pred=LR.predict(Xtest)
    rmse.append(round(np.sqrt(metrics.mean_squared_error(Ytest,Y_pred)),2))
    r2.append(round(LR.score(Xtest,Ytest),2))
```
