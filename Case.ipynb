```python
def f(x):
    b=dict(ReduceStreetNoise=x[x['Benefits']=='Reduce Street Noise']['Benefits'].count(),
                       Nil=x[x['Benefits']=='Nil']['Benefits'].count(),
           ReduceACEnergyCostStreetNoise=x[x['Benefits']=='Reduce AC Energy Cost & Street Noise']['Benefits'].count(), 
              
           AestheticsLowMaintenance=x[x['Benefits']=='Aesthetics & Low Maintenance']['Benefits'].count(),             
           LowMaintenance=x[x['Benefits']=='Low Maintenance']['Benefits'].count(),                           
           AestheticsStreetNoise=x[x['Benefits']=='Aesthetics & Street Noise']['Benefits'].count(),                
           ReduceStreetNoiseLowMaintenance=x[x['Benefits']=='Reduce Street Noise & Low Maintenance']['Benefits'].count(),    
           NoiseProofBetterLighting=x[x['Benefits']=='Noise Proof & Better Lighting']['Benefits'].count(),             
           LowMaintenanceAesthetics=x[x['Benefits']=='Low Maintenance & Aesthetics']['Benefits'].count())
    return Series(b)
```
